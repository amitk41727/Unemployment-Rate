```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
#from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
data.head()
```

data= data.rename(columns={data.columns[0]:'State',data.columns[3]:'EUR',data.columns[4]:'EE', data.columns[5]:'ELPR',data.columns[6]:'Region'})

```python
plt.style.use("seaborn-darkgrid")
plt.figure(figsize=(8,6))
heat_maps= data[['EUR','EE', 'ELPR','longitude', 'latitude']]
sns.heatmap(heat_maps.corr(),cmap='gist_gray')
plt.show()
```

```python
plt.figure(figsize=(9,7))
plt.title("Unemployment in India")
sns.histplot(x="EUR",hue="Region",data=data,element="bars")
plt.show()
```

```python
UnEmp=data[["State","Region","EUR"]]
figure=px.sunburst(UnEmp,path=["Region","State"],values="EUR",width=550,height=550,title="Unemployment Rate In India")
figure.show()
```
